```python
import pandas as pd

# Step 1: Load the dataset
file_path = 'data_bet.csv'
data = pd.read_csv(file_path)

# Step 2: Filter relevant columns
relevant_data = data[['league', 'home_team', 'guest_team', '2nd_half_under']]

# Step 3: Calculate probabilities for leagues, home teams, and guest teams
# Probability of 2nd half under for each league
league_probabilities = relevant_data.groupby('league')['2nd_half_under'].apply(lambda x: (x == -1).mean()).sort_values(ascending=False)

# Probability of 2nd half under for each home team
home_team_probabilities = relevant_data.groupby('home_team')['2nd_half_under'].apply(lambda x: (x == -1).mean()).sort_values(ascending=False)

# Probability of 2nd half under for each guest team
guest_team_probabilities = relevant_data.groupby('guest_team')['2nd_half_under'].apply(lambda x: (x == -1).mean()).sort_values(ascending=False)

# Step 4: Display results
print("Leagues with Highest Probability of 2nd Half Under:")
print(league_probabilities)

print("\nHome Teams with Highest Probability of 2nd Half Under:")
print(home_team_probabilities)

print("\nGuest Teams with Highest Probability of 2nd Half Under:")
print(guest_team_probabilities)
```

```python
# Save the probabilities to CSV files
league_probabilities.to_csv('league_probabilities.csv', header=['probability'])
home_team_probabilities.to_csv('home_team_probabilities.csv', header=['probability'])
guest_team_probabilities.to_csv('guest_team_probabilities.csv', header=['probability'])

print("Results have been successfully saved to:")
print("1. league_probabilities.csv")
print("2. home_team_probabilities.csv")
print("3. guest_team_probabilities.csv")
```
